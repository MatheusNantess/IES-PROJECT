```python
#importing the libraries and giving the path to the file of data base and giving the right diretory 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import unicodedata 
import re
new_directory = 'C:/Users/mathe/OneDrive/Pictures/Python/CsvExcelFiles'
os.chdir(new_directory)

```

```python
#reading the data set and setting only the five rows
df = pd.read_csv('IES.csv')
df.head(5)
```

```python
#Columns and dtype
print("Numero de Colunas:")
print(df.columns)
```

```python
#removing NaN
df.dropna()
```

```python
#when you use df.columns, it returns an array which can be iterated 
#here I removed all the useless points  in portuguese that we have
def remover_acentos(txt):
    txt = unicodedata.normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII')
    return re.sub('[^a-zA-Z0-9 \\\]', '', txt)

for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].apply(lambda x: remover_acentos(x) if isinstance(x,str) else x)
```

#organize the amount of teachers  by age group, color, nationality through the years


```python
#veryfing the columns with 'quantidade' and 'docente'
colunas_demograficas = [coluna for coluna in df.columns if 'quantidade' in coluna and 'docente' in coluna]
colunas_demograficas

```

```python
#quantity of teacher by age through the years
#here i separated a group of docents by age in every year
#then i transformed the data i had into a data with only three columns to plot what i want
#i also replaced the string to better visualization 
amount_teacher_age = df.groupby('ano')[['quantidade_docentes_exercicio_60_mais'
,'quantidade_docentes_exercicio_55_59'
,'quantidade_docentes_exercicio_50_54'
,'quantidade_docentes_exercicio_45_49'
,'quantidade_docentes_exercicio_40_44'
,'quantidade_docentes_exercicio_35_39'
,'quantidade_docentes_exercicio_30_34'
,'quantidade_docentes_exercicio_0_29']].sum()
amount_teacher_age_reset = amount_teacher_age.reset_index().melt(id_vars='ano',var_name='faixa_etaria',value_name='quantidades')
amount_teacher_age_reset['faixa_etaria'] =  amount_teacher_age_reset['faixa_etaria'].str.replace('quantidade_docentes_exercicio_','')
amount_teacher_age_reset['faixa_etaria'] =  amount_teacher_age_reset['faixa_etaria'].str.replace('_','-')
amount_teacher_age_reset
```

```python
#plotting the amount by age 
plt.figure(figsize=(14,8))

sns.lineplot(data=amount_teacher_age_reset,x='ano',y='quantidades',hue='faixa_etaria', marker='o')
plt.title('Quantidade de Professores por idade ao longo dos anos')
plt.ylabel('Quantidade de Professores')
plt.xlabel('Anos')

plt.tight_layout()
```

```python
#getting the amount by nationality 
#first I used groupby to select groups by year, and then located all columns that I want to sum
#then I reseted the index to use melt, to transform the data for better visualization
#then I cleaned the values in the columns to make it easier to plot
amount_teacher_nationality = df.groupby('ano')[['quantidade_docentes_exercicio_estrangeiro','quantidade_docentes_exercicio_brasileiro']].sum()
amount_teacher_nationality_reset = amount_teacher_nationality.reset_index().melt(id_vars='ano',var_name='nacionalidade',value_name='quantidade')
amount_teacher_nationality_reset['nacionalidade'] = amount_teacher_nationality_reset['nacionalidade'].str.replace('quantidade_docentes_exercicio_','')
amount_teacher_nationality_reset
```

```python
#building the graphic based on the melted dtaframe
plt.figure(figsize=(14,8))
sns.lineplot(data=amount_teacher_nationality_reset, x='ano',y='quantidade',hue='nacionalidade',marker='o')
plt.xlabel('Anos')
plt.ylabel('Amount of teachers')
plt.title('Amount of teachers by nationatlity through the years')

```

```python
#amount of teachers by race
#first I used groupby to select groups by year, and then located all columns that I want to sum
#then I reseted the index to use melt, to transform the data for better visualization
#then I cleaned the values in the columns to make it easier to plot
amount_teacher_race = df.groupby('ano')[['quantidade_docentes_exercicio_branca',
       'quantidade_docentes_exercicio_preta',
       'quantidade_docentes_exercicio_parda',
       'quantidade_docentes_exercicio_amarela',
       'quantidade_docentes_exercicio_indigena',
       'quantidade_docentes_exercicio_cor_nao_declarada',]].sum()
amount_teacher_race_reset = amount_teacher_race.reset_index().melt(id_vars='ano',var_name='race',value_name='amount')
amount_teacher_race_reset['race'] = amount_teacher_race_reset['race'].str.replace('quantidade_docentes_exercicio_','')
amount_teacher_race_reset

```

```python
#ploting the amount by race through years
plt.figure(figsize=(14,8))
sns.lineplot(data=amount_teacher_race_reset,x='ano',y='amount',hue='race',marker='o')
plt.title("Amount of Teachers by race Through Years")
plt.ylabel("Amount of Teachers")
plt.xlabel("Years")

plt.tight_layout()

```

#Is there any significant difference between the ies in different states in the teachers composition?

#1 - by gender (only exists male/famale)


```python
#grouping teachers composition by gender
#first I did the name as before, but this time i had to deL the columns aftewards to show the plot 
gender_amount_states = df.groupby('sigla_uf')[[
        'quantidade_docentes_exercicio_feminino',
       'quantidade_docentes_exercicio_masculino',]].sum()
#creating a column based in the sum of the other two gender columns
gender_amount_states['total_docentes'] = gender_amount_states.sum(axis=1)
#calculating the proportion based on the new total column i created
gender_amount_states['proporcao_feminino'] = gender_amount_states['quantidade_docentes_exercicio_feminino'] / gender_amount_states['total_docentes']
gender_amount_states['proporcao_masculina'] = gender_amount_states['quantidade_docentes_exercicio_masculino'] / gender_amount_states['total_docentes']
#here i just separated the proportion column from the other that I wont use
gender_amount_states_prop = gender_amount_states[['proporcao_feminino','proporcao_masculina']]
#
gender_amount_states_prop_reset = gender_amount_states_prop.reset_index().melt(id_vars='sigla_uf',var_name='gender',value_name='Proportion')
gender_amount_states_prop_reset['gender'] = gender_amount_states_prop_reset['gender'].str.replace('proporcao_','')
gender_amount_states_prop_reset.head(5)
```

```python
#ploting the graphic, i made a barplot with the collors representing respectives genders
plt.figure(figsize=(14,8))
palette_colors = {"masculina": "blue", "feminino": "#c071d0"}
sns.barplot(data=gender_amount_states_prop_reset,x='sigla_uf',y='Proportion',hue='gender', palette=palette_colors)
plt.title('Proportion of Teachers by FU (Federal Unities)')
plt.xlabel('Federal Unities')
plt.ylabel('Proportion')
```

#Teachers with disabilities: What is the proportion of teachers with disabilities in the top ten IES wi this teachers? Are there trends of growth or decline over the years?

```python
#group all disabilities teachers by ies
disabilities_ies = df.groupby('nome_mantenedora')[['quantidade_docentes_exercicio_deficiencia','quantidade_docentes_exercicio']].sum()

#renaming the column
disabilities_ies.columns = ['deficient_teachers','total_teachers']
top_10_disabilities_ies  = disabilities_ies.sort_values(by='deficient_teachers',ascending=False).head(10)
#calculating the proportion base on a total number
top_10_disabilities_ies['deficient_teachers_prop'] = top_10_disabilities_ies['deficient_teachers'] / top_10_disabilities_ies['total_teachers']
#here i cut the data frame to the only ten IES with most teachers with desabilities, now im doing a proper comparision
top_10_disabilities_ies['deficient_teachers_prop'] = top_10_disabilities_ies['deficient_teachers_prop']*100


#in this block of code, i will see the amount of disabilities teachers over the year
disabilities_ies_year = df.groupby('ano')[['quantidade_docentes_exercicio_deficiencia','quantidade_docentes_exercicio']].sum()
disabilities_ies_year.columns = ['deficient_teachers','total_amount']
disabilities_ies_year.reset_index()
```

```python
#ploting the graphic of the Percentege of Teachers with desabilities related to the Total Amount
import matplotlib.ticker as mtick
#plot to show the percentage
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
sns.barplot(data=top_10_disabilities_ies,x='deficient_teachers_prop',y='nome_mantenedora',hue='nome_mantenedora')
plt.xlabel('percentage of teachers with disabilities out of the total number of Teachers')
#code to add a %symbol in the graphic
plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(100))
plt.title('Percentege of Teachers with desabilities related to the Total Amount')
plt.ylabel('IES')

#plot to show the number of disabilities professor over the years
plt.subplot(1,2,2)
sns.lineplot(data=disabilities_ies_year, y='deficient_teachers',x='ano')
plt.xlabel('year')
```


#Racial Diversity among Regions: How does racial diversity among Regions vary? 

```python
#forst I am groupig by states all races
race_teachers = df.groupby('sigla_uf')[['quantidade_docentes_exercicio_branca',
       'quantidade_docentes_exercicio_preta',
       'quantidade_docentes_exercicio_parda',
       'quantidade_docentes_exercicio_amarela',
       'quantidade_docentes_exercicio_indigena',]].sum()
race_teachers.columns = ['Branca', 'Preta','Parda','Amarela','Indigena']
#here i am summing all values to know the total numbers 
race_teachers['total_race'] = race_teachers.sum(axis=1)

#lets create the columns of percentage
race_teachers['Branca %'] = race_teachers['Branca'] / race_teachers['total_race']
race_teachers['Amarela %'] = race_teachers['Amarela'] / race_teachers['total_race']
race_teachers['Preta %'] = race_teachers['Preta'] / race_teachers['total_race']
race_teachers['Indigena %'] = race_teachers['Indigena'] / race_teachers['total_race']
race_teachers['Parda %'] = race_teachers['Parda'] / race_teachers['total_race']


#preparing the data to ploting, selecting only the proportion part and creplacing the ' %' string
race_teachers_prop = race_teachers[['Branca %','Amarela %','Preta %','Indigena %','Parda %']]
race_teachers_prop_long = race_teachers_prop.reset_index().melt(id_vars='sigla_uf',var_name='race',value_name='Proportion')
race_teachers_prop_long['race'] = race_teachers_prop_long['race'].str.replace(' %','')
race_teachers_prop_long
```

```python
#now i will plot the graphic 
#Exporting to put % in the xlabel
import matplotlib.ticker as mtick
plt.figure(figsize=(14,8))
sns.barplot(data=race_teachers_prop_long,x='Proportion',y='sigla_uf',hue='race')
#this command is to just set the xlabel a % symbol
plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1))
```
